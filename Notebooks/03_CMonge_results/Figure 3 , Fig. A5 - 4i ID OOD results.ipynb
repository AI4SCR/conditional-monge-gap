# All results

```python
res = pd.read_csv("/Users/alicedriessen/Library/CloudStorage/Box-Box/otperturb/alice_exp/models/4i_ood/mean_results.csv", index_col=0)
res["drug_combi"] = [len(d.split("_"))>1 for d in res["drug"]]
res = res[res["eval_setting"]!="homogeneous"]

homogeneous_res = pd.read_csv("/Users/alicedriessen/Library/CloudStorage/Box-Box/otperturb/alice_exp/models/4i_id/homogeneous_mean_results.csv", index_col=0)
h_res = homogeneous_res[homogeneous_res["eval_setting"]=="Cmonge-homogeneous"]

monge_ood_res = pd.read_csv("/Users/alicedriessen/Library/CloudStorage/Box-Box/otperturb/alice_exp/models/4i_ood/monge_mean_results.csv", index_col=0)

res = pd.concat([res, h_res, monge_ood_res, monge_id_res])
res[~res["drug"].isin(["vemurafenib_cobimetinib", "pomalidomide_carfilzomib_dexamethasone"])]
res[~res["ood_drug"].isin(["vemurafenib_cobimetinib", "pomalidomide_carfilzomib_dexamethasone"])]
res["drug_combi"] = [len(d.split("_"))>1 for d in res["drug"]]

res["Model"] = None
res["split"] = None
res.loc[(res["embedding"]=="moa")&(res["eval_setting"]=="CMonge-ood"), ["Model", "split"]] = ("CMonge-MoA", "OOD")
res.loc[(res["embedding"]=="moa")&(res["eval_setting"]=="Cmonge-homogeneous"), ["Model", "split"]] = ("CMonge-MoA", "ID")

res.loc[(res["embedding"]=="rdkit")&(res["eval_setting"]=="CMonge-ood"), ["Model", "split"]] = ("CMonge-RDkit", "OOD")
res.loc[(res["embedding"]=="rdkit")&(res["eval_setting"]=="Cmonge-homogeneous"), ["Model", "split"]] = ("CMonge-RDkit", "ID")

res.loc[res["eval_setting"]=="Monge-id", ["Model", "split"]] = ("Monge-IS", "ID")
res.loc[res["eval_setting"]=="Monge-ood", ["Model", "split"]] = ("Monge-OOS", "OOD")
res.loc[res["eval_setting"]=="within condition", "Model"] = "Within condition"

within_condition_res = res[res["Model"]=="Within condition"]

res = res[res["eval_setting"]!="within condition"]
res = res[res["eval_setting"]!="Identity"]
res = res[res["eval_setting"]!="identity"]
```

# In sample results

```python
monge_res = monge_res.drop_duplicates()
identity_res = identity_res.drop_duplicates()

monge_res["drug_combi"] = [len(d.split("_"))>1 for d in monge_res["drug"]]
identity_res["drug_combi"] = [len(d.split("_"))>1 for d in identity_res["drug"]]

mean_monge_r2 = monge_res["mean_r2"].mean()
mean_monge_single_r2 = monge_res.loc[~monge_res["drug_combi"], "mean_r2"].mean()
mean_monge_combi_r2 = monge_res.loc[monge_res["drug_combi"], "mean_r2"].mean()

mean_monge_emd = monge_res["mean_wasserstein"].mean()
mean_monge_single_emd = monge_res.loc[~monge_res["drug_combi"], "mean_wasserstein"].mean()
mean_monge_combi_emd = monge_res.loc[monge_res["drug_combi"], "mean_wasserstein"].mean()

mean_identity_r2 = identity_res["mean_r2"].mean()
mean_identity_single_r2 = identity_res.loc[~identity_res["drug_combi"], "mean_r2"].mean()
mean_identity_combi_r2 = identity_res.loc[identity_res["drug_combi"], "mean_r2"].mean()

mean_identity_emd = identity_res["mean_wasserstein"].mean()
mean_identity_single_emd = identity_res.loc[~identity_res["drug_combi"], "mean_wasserstein"].mean()
mean_identity_combi_emd = identity_res.loc[identity_res["drug_combi"], "mean_wasserstein"].mean()

mean_identity_mmd = identity_res["mean_mmd"].mean()
mean_identity_single_mmd = identity_res.loc[~identity_res["drug_combi"], "mean_mmd"].mean()
mean_identity_combi_mmd = identity_res.loc[identity_res["drug_combi"], "mean_mmd"].mean()

mean_within_condition_r2 = within_condition_res["mean_r2"].mean()
mean_within_condition_single_r2 = within_condition_res.loc[~within_condition_res["drug_combi"], "mean_r2"].mean()
mean_within_condition_combi_r2 = within_condition_res.loc[within_condition_res["drug_combi"], "mean_r2"].mean()

mean_within_condition_emd = within_condition_res["mean_wasserstein"].mean()
mean_within_condition_single_emd = within_condition_res.loc[~within_condition_res["drug_combi"], "mean_wasserstein"].mean()
mean_within_condition_combi_emd = within_condition_res.loc[within_condition_res["drug_combi"], "mean_wasserstein"].mean()

mean_within_condition_mmd = within_condition_res["mean_mmd"].mean()
mean_within_condition_single_mmd = within_condition_res.loc[~within_condition_res["drug_combi"], "mean_mmd"].mean()
mean_within_condition_combi_mmd = within_condition_res.loc[within_condition_res["drug_combi"], "mean_mmd"].mean()
```
