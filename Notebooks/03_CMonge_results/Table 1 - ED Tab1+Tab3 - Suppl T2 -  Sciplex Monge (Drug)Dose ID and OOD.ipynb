```python
import pandas as pd
import yaml
```

# Monge DrugDose OOD

```python
drugs = ["abexinostat", "belinostat", "dacinostat", "entinostat", "givinostat", 
             "mocetinostat", "pracinostat", "tacedinaline", "trametinib"]
exp_order = ["homogenous_valid_set"]+ ["Monge-ood", "Identity", "within_condition"] * 4
dosages=["-10", "-100", "-1000", "-10000"]

all_mean_metrics = []
for drug in drugs:

    with open(f"/Users/adr/Box/otperturb/alice_exp/sciplex_monge_ood/{drug}/logs.yaml") as stream:
        logs = yaml.safe_load(stream)

    print(drug, len(logs["experiments"]))
    for i, d in enumerate(logs["experiments"]):
        if i==0:
            continue
            
        mean_stats = d["mean_statistics"]
        mean_stats = pd.DataFrame.from_dict(mean_stats, orient="index").T
        mean_stats["eval_setting"] = exp_order[i]
        mean_stats["ood_drug"] = drug
        mean_stats["model"] = "Monge-ood"
        mean_stats["condition"] = drug+dosages[(i-1)%4]
        all_mean_metrics.append(mean_stats)
```

```python
ood_res = pd.concat(all_mean_metrics)
ood_res[["drug", "dose"]] = [c.split("-") for c in ood_res["condition"]]
ood_res = ood_res[ood_res["eval_setting"]=="Monge-ood"].reset_index(drop=True)
```

```python
ood_res.groupby(["dose"]).mean(numeric_only=True)
```

```python
ood_res.groupby(["dose"]).std(numeric_only=True)
```

# Monge Dose OOD

```python
drugs = ["abexinostat", "belinostat", "dacinostat", "entinostat", "givinostat", 
             "mocetinostat", "pracinostat", "tacedinaline", "trametinib"]
exp_order = ["homogenous_valid_set"]+ ["Monge-ood", "Identity", "within_condition"]
dosages=["-10", "-100", "-1000", "-10000"]

all_mean_metrics = []
for drug in drugs:
    for dose in dosages:

        with open(f"/Users/adr/Box/otperturb/alice_exp/sciplex_monge_dose_ood/{drug}{dose}/logs.yaml") as stream:
            logs = yaml.safe_load(stream)
    
        print(drug, len(logs["experiments"]))
        for i, d in enumerate(logs["experiments"]):
        
            mean_stats = d["mean_statistics"]
            mean_stats = pd.DataFrame.from_dict(mean_stats, orient="index").T
            mean_stats["eval_setting"] = exp_order[i]
            mean_stats["ood_drug"] = f"{drug}{dose}"
            mean_stats["model"] = "Monge-ood"
            mean_stats["drug"] = drug
            mean_stats["dose"] = dose[1:]
            all_mean_metrics.append(mean_stats)
```

```python
dose_ood = pd.concat(all_mean_metrics).reset_index(drop=True)
```

```python
dose_ood = dose_ood[dose_ood["eval_setting"]=="Monge-ood"].reset_index(drop=True)
dose_ood.groupby("dose").mean(numeric_only=True)
```

```python
dose_ood.groupby("dose").std(numeric_only=True)
```
